```python
import control as ctrl
import matplotlib.pyplot as plt

num = [1]
den = [1,1,1]
G = ctrl.tf(num, den)
print (G)
```

```python
poles = ctrl.poles(G)
print("Poles:", poles)
```

```python
zeros = ctrl.zeros(G)
print("Zeros:", zeros)
```

```python
import matplotlib.pyplot as plt

ctrl.pzmap(G, grid=True)
plt.show()
```
